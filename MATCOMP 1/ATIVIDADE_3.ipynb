<font color='#079ca3'> **Derivada numerica com h = 0.1, intervalo [1,3] e com 21 pontos:**

---

```python
import numpy as np
import matplotlib.pyplot as plt

xi = np.linspace(1, 3, 21)
f = lambda x: np.exp(x / 3) + 1 # fun√ß√£o
h = 0.1
x = 2

frente = (f(x + h) - f(x)) / h
tras = (f(x) - f(x - h)) / h
centro = (f(x + h) - f(x - h)) / (2 * h)
analitica = lambda x: np.exp(x / 3) / 3

print('Analitica: ', analitica(2))
print("Diferen√ßa para Frente:", frente)
print("Diferen√ßa para Tr√°s:", tras)
print("Diferen√ßa para o centro", centro)
```

<font color='#079ca3'> **Derivada numerica com h = 0.05, intervalo [1,3] e com 41 pontos:**

---

```python
import numpy as np
import matplotlib.pyplot as plt

xi_2 = np.linspace(1, 3, 41)
f = lambda x: np.exp(x / 3) + 1 # fun√ß√£o
h_2 = 0.05
x = 2

frente_2 = (f(x + h_2) - f(x)) / h_2
tras_2 = (f(x) - f(x - h_2)) / h_2
centro_2 = (f(x + h_2) - f(x - h_2)) / (2 * h_2)
analitica = lambda x: np.exp(x / 3) / 3
print('Analitica: ', analitica(2))
print("Diferen√ßa para Frente:", frente_2)
print("Diferen√ßa para Tr√°s:", tras_2)
print("Diferen√ßa para o centro", centro_2)
```

```python
# 3/8 de Simpson
x0 = - np.pi/2
nx = np.pi/2
f = lambda x: (4 * np.cos(x) / (3 + 2 * np.sin(x)))
for n in  [4, 30, 60]:
  h = (nx - x0) / n
  x = np.linspace(x0, nx, n + 1)
  y = f(x)

  I = (3 * h / 8) * (y[0] + 3 * np.sum(y[1:n - 1:3] + y[2:n:3]) + 2 * np.sum(y[3:n:3]) + y[-1])
  print('| n = ', n, ' I =', I, ' |')
```
