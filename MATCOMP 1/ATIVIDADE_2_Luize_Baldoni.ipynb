Encontrando funo derivada para aplicar no metodo de newton

```python
import sympy as sp
t = sp.symbols('t')
f = - (10 / 4) * (2 * t ** 2 - sp.cos(2 * t)) # funo principal
sp.diff(f) #derivando
```

```python
import numpy as np
f = lambda t: - (10 / 4) * (2 * t ** 2 - np.cos(2 * t)) # Definio da funo
df = lambda t: - 10.0 * t - 5.0 * np.sin(2 * t) # Definio da funo derivada

t = 0.5 # t inicial
t_ant = t
err = 1.0 # erro inicial
convergencia = 0 # contagem de convergencias

#metodo de newton
while err > (10 ** -6):
  t = t - f(t) / df(t)
  err = abs(t - t_ant)
  t_ant = t
  convergencia += 1
  print(" | tempo = %.9f"%t, " | erro = %.9f"%err, '| convergencia = ', convergencia, '|')
```

<font color='#FF69B4'> **OBSERVAO:** </font>
Para um tempo aproximado (t) o objeto atinge o solo, ou seja com o tempo 0.510844977 o objeto atinge o solo

### <font color='#FF69B4'> **MTODO DA BISSEO**


---




```python
def bissec(f, a, b, eps = (10 ** -6)):
    err = b - a
    t_ant = a
    convergencia2 = 0 # contagem de convergencias

    while err > eps:
        t = (a + b) / 2.0
        if f(t) == 0:
            break
        elif f(t) * f(a) < 0:
            b = t
        else:
            a = t
        err = abs(t - t_ant)
        t_ant = t
        convergencia2 += 1 # adiconando 1 a contagem
        print ('| t = %.9f'%t, '| err = %.9f'%err, '| convergencia = ', convergencia2, '|')

    return (t)
f = lambda t: - (10 / 4) * (2 * t ** 2 - np.cos(2 * t)) #funao principal
print('|  bissec = ', bissec(f, 0.3, 0.6, 0.000001), '|')

```

<font color='#FF69B4'> **OBSERVAO:** </font>
Para um tempo aproximado (t) o objeto atinge o solo, ou seja com o tempo 0.510845089 o objeto atinge o solo e tendo uma bisseo de 0.5108450889587401

### <font color='#FF69B4'> **MTODO DAS SECANTES**


---



```python
import numpy as np
err = 1.0 # erro inicial
t0 = 0.5 # valor de t incial
t1 = 0.7 # valor de t1 inicial
convergencia3 = 0 # contar quantas convergencias

f = lambda t: - (10 / 4) * (2 * t ** 2 - np.cos(2 * t))

while err > (10 ** -6):
  t = ( t0 * f(t1) - t1 * f(t0)) / (f(t1) - f(t0))
  err= abs(t - t1)
  t0 = t1
  t1 = t
  convergencia3 += 1 # adicionando 1 a contagem
  print("| t = %.9f"%t, "| err = %.9f"%err, '| convergencia = ', convergencia3, '|')
```

<font color='#FF69B4'> **OBSERVAO:** </font>
Para um tempo aproximado (t) o objeto atinge o solo, ou seja com o tempo 0.510844977 o objeto atinge o solo

O metodo que sera utilizado  o de Newton devido ao fato dele ser mais rapido e de melhor compreendimento portanto, vamos encontrar a derivada da funo para aplicarmos no metodo posteriormente:

---



```python
import sympy as sp
h = sp.symbols('h')
f = (10 * (1 / 2 * (np.pi * (1 ** 2)) - ((1 ** 2) * sp.asin(h / 1)) - (h * ((1 ** 2) - (h ** 2)) ** (1 / 2))) ) - 12.4
print(sp.diff(f)) #derivando
```

<font color='#FF69B4'> **OBSERVAO:** </font> foi subtraido 12.4 da funo f devido ao fato que a questao ja nos da o valor do volume, entao para descobrirmos o volume devemos subtrair o valor do mesmo



---

<font color='#FF69B4'> **MTODO DE NEWTON E TABELA DE REPORTE**




```python
import numpy as np
f = lambda h:(10 * (1 / 2 * (np.pi * (1 ** 2)) - ((1 ** 2) * sp.asin(h / 1)) - (h * ((1 ** 2) - (h ** 2)) ** (1 / 2))) ) - 12.4 # Definio da funo
df = lambda h: 10.0 * h ** 2 / (1 - h**2) ** 0.5 - 10 * (1 - h**2) ** 0.5 - (10 / (1 - h**2) ** 1 / 2) # Definio da funo derivada

h = 0 # h inicial
h_ant = h
err = 1.0 # erro inicial
convergencia = 0 # contagem de convergencias

# comeando tabela de reporte
print('|    ITERAO\tAPROXIMAO\t   f(hk)\t   ERRO\t\tPROFUNDIDADE     |')

#metodo de newton
while err > 0.01:
  h = h - f(h) / df(h)
  err = abs(h - h_ant)
  h_ant = h
  volume = (10 * (1 / 2 * (np.pi * (1 ** 2)) - ((1 ** 2) * sp.asin(h / 1)) - (h * ((1 ** 2) - (h ** 2)) ** (1 / 2))) )  # definindo volume para cada valor
  profundidade = 1 - h # profundidade em cada valor
  convergencia += 1

  #REPORTANDO NUMEROS DE ITERAES, APROXIMAO, ERRO, VALOR DE F(hk) E PROFUNDIDADE
  print('|--------------------------------------------------------------------------------|')
  print('|      %d \t%.9f \t%.9f \t%.9f \t%.9f      |' % (convergencia, h , volume, err, profundidade))
print('|--------------------------------------------------------------------------------|')
```

<font color='#FF69B4'> **PROFUNDIDADE:** </font>
Como calculado atravez do metodo de newton, pode-se ver que a profundidade  aproximadamente <font color='#FF69B4'>**0.836026909**
