**Nos proximos trechos de codigo sera descoberto o valor de Y para equação Cubica**

Sao definidos os valores de <font color="green">**L0, L1, L2 e L3**</font>
para conseguir achar qual a equação de y a partir dos pontos pre-definidos atraves da tabela dada pelo enunciado do exercicio

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

x, u = symbols("x, u", real=True)
'''pontos equação cubica'''
Xi = [0.01, 0.02, 0.03, 0.04] # pontos de X
Yi = [4.0379, 3.3547, 2.9591, 2.6813] # pontos de Y
def calculaL(k, Xi):
    Lk = 1
    Lk_num = 1
    Lk_den = 1
    # laço de repetição para gerar as equações
    for i, xi in enumerate(Xi):
      xk = Xi[k]
      if i!=k:
        Lk_num = Lk_num * (x-xi)
        Lk_den = Lk_den * (xk-xi)
    return expand(Lk_num)/Lk_den # gera os valores de L
'''
    DESCOBRINDO VALORES DE L EM P
'''
def calculaP(Xi):
    P = 0
    for k in range(len(Xi)):
      Lk = calculaL(k, Xi)
      print ('L',k,'=',Lk)
      P = P + Yi[k]*Lk
    return P
P = calculaP(Xi) # gera os valores de L
```



---

Os proximos trechos de codigo irao calcular o polinomio interpolador de Newton de f(x) = e^2x com xE[-1, 1] e com <font color='green'> **n +1 = 5**<font>

```python
import numpy as np
from sympy import*
import matplotlib.pyplot as plt
from matplotlib import style

func = lambda x: np.exp(2*x) # definindo a funçao de y
xi = np.linspace(-1, 1, 5)  # definindo 5 pontos
yi = func(xi)

# difereças divididas
def difdiv(x,y):
    m = len(x)
    dd = y.copy()
    for k in range(1,m):
        dd[k:m]=(dd[k:m]-dd[k-1])/(x[k:m]-x[k-1])
    return dd
f = difdiv(xi,yi)
print(f)

def calculaP(f,xi,x):
    n = len(xi) - 1
    p = f[n]
    for i in range(1,n+1):
        p = f[n-i] + (x -xi[n-i])*p
    return p

# Cria pontos e calcula os valores para o plot
xp = np.linspace(-1, 1)
yp = []
for x in xp:
    y = calculaP(f,xi,x)
    yp.append(y)
```
